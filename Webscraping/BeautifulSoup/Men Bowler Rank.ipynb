```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
df_bowl = pd.DataFrame(columns = ['Rank', 'Name', 'Nation', 'Rating'])

rank1 = soup.find('span', class_='rankings-block__pos-number').text.replace('\n','').strip(' ')
name1 = soup.find('div', class_='rankings-block__banner--name-large').text
nation1 = soup.find('div', class_='rankings-block__banner--nationality').text.replace('\n','')
rating1 = soup.find('div', class_='rankings-block__banner--rating').text

df_bowl = df_bowl.append([{'Rank':rank1,
                           'Name':name1,
                           'Nation':nation1,
                           'Rating':rating1}],ignore_index=True)

players = soup.find('table', class_='table rankings-table').find_all('tr', class_='table-body')

for player in players:
    rank = player.find('td', class_='table-body__cell table-body__cell--position u-text-right').div.span.text.replace('\n','').strip(' ')
    name = player.find('td', class_='table-body__cell rankings-table__name name').find('a').text
    nation = player.find('td', class_='table-body__cell nationality-logo rankings-table__team').text.replace('\n','').strip(' ')
    rating = player.find('td', class_='table-body__cell rating').text
    df_bowl = df_bowl.append([{'Rank':rank,
                               'Name':name,
                               'Nation':nation,
                               'Rating':rating}],ignore_index=True)
df_bowl
```

```python
new_df_bowl = df_bowl.head(10)

new_df_bowl.to_csv('ICC Top 10 Bowler(Men).csv', index = False)
```
