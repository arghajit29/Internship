```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://presidentofindia.nic.in/former-presidents.htm"
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
df = pd.DataFrame(columns = ['Former President Name','Term of Office'])
pres = soup.find('ul', class_='listing cf').find_all('li')
for p in pres:
    name = p.find('div', class_='presidentListing').find('h3').text
    term = p.find('div', class_='presidentListing').find('p').text.replace('Term of Office: ','')
    df = df.append([{'Former President Name': name,
                     'Term of Office': term}],ignore_index=True)
    
df
```

```python
df.to_csv('Former Presidents.csv', index=False)
```
